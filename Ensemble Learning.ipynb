```python
#Q5.  How can you measure the importance of features in a Random Forest model?
"""
Ans5. In a random forest model, feature importance helps determine which feature contribute the most to 
      predictions. There aer several ways to measure it:

      1. Gini Importance (Mean Decrease in Impurity)
          - Each tree in the forest splits data based on feature.
          - The Gini Impurity measures how mixed the data is at each split.
          - Features that reduce impurity the most are considered more important.
      
      2. Permutation Importance
          - Randomly shuffle a features value and checks how much the model accuracy drops.
          - If accuracy drops significantly, the feature is important.
      3. SHAP (SHapley Additive exPlanations Values)
          - Provides a detailed breakdown of how each feature impacts individual predictions.
          - Helps interpret the model's decision-making process.
      
"""
```

```python
#Q6. Explain the working principle of a Bagging Classifier.
"""
Ans6. 1. Bootstrap Smapling:
            The training dataset is randomly samples with replacement to create multiple subsets. Some data point
            may appear multiple times, while other may be left out.
      2. Base model training:
            Each subset is used to train a seaparate model (e.g., decision trees, SVM, etc) independently and in 
            parallel.
      3. Aggregation:
            The predictions from all models are combined using majority voting(for classification) and averaging
            (for regression).
      4. Final Prediction:
            The combined result is more accurate and stable than a single model, reducing variance and improve
            generalization.
"""
```

```python
#Q13.  What are the main types of ensemble techniques?
"""
Ans13. The main type of ensemble techniques are:
        1. Bagging (Bootstrapping Aggregating)
            - Trains multiple models independently on random subset of data.
            - Reduces variance and prevents overfitiing
            - Example: Random Forest
        2. Boosting
            - Models are train sequentially, with each new model focusing on correcting errors made by previous one.
            - Improves accuracy by reducing bias.
            - Example: AdaBoost, Gradient Boodsting, XGBoost
        3. Stacking
            - Combines different models nad uses a meto model to refine predictions.
            - Learns how to best combine the outputs of base models.
            - Example: Stacked Ensemble Models.
"""
```

```python
#Q45. Train a Bagging Regressor with different levels of bootstrap samples and compare performance.
```

```python
X,y = make_regression(n_samples=1000, n_features=10, noise=0.1, random_state=1)
```

```python
#train test split
X_train,X_test, y_train,y_test = train_test_split(X, y , test_size=0.2, random_state=1)
```

```python
bootstrap_smaples = [0.5, 0.7, 1.0]
```

```python
for i in bootstrap_smaples:
    bg_reg = BaggingRegressor(estimator=DecisionTreeRegressor(),
                    n_estimators=10,
                    max_samples=i,
                    random_state=1)
    bg_reg.fit(X_train, y_train)
    y_pred = bg_reg.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    print(f"Bagging regressor with {i} bootstarp samples - MSE: {mse}")
```
